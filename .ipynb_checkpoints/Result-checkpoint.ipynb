```python
import requests
import pandas as pd
import time
import datetime
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#### Create DF

```python
df = pd.read_csv('netflix_combined.csv')
```

```python
df.head(5)
```

#### As the predictor variable is 'SubscriptionStatus' and it only has two category (active or not active), I need to convert the data to binary first

```python
def convert_status_to_binary(status):
    if status == 'Active':
        return 1
    else:
        return 0
```

```python
df['SubscriptionStatus'] = df['SubscriptionStatus'].apply(convert_status_to_binary)
```

```python
df.head(5)
```

#### Before proceeding forward, I have to convert Gender, BirthDate, Country, and Payment Method first. The rest I will figure out after.

##### Converting gender with one-hot encoder

```python
encoder = OneHotEncoder()
encoder.fit(df[['Gender']])
gender_encoded = encoder.transform(df[['Gender']])
gender_encoded_df = pd.DataFrame(gender_encoded.toarray(), columns=encoder.get_feature_names(['Gender']))
df = pd.concat([df, gender_encoded_df], axis=1)
```

```python
df.drop('Gender', axis=1, inplace=True)
```

```python
df.head(5)
```

##### Converting BirthDate into three columns: Year, Month, Day

```python
df[['BirthYear', 'BirthMonth', 'BirthDay']] = df['BirthDate'].str.split('-', expand=True)
df.drop('BirthDate', axis=1, inplace=True)
```

```python
df['BirthYear'] = df['BirthYear'].astype(int)
df['BirthMonth'] = df['BirthMonth'].astype(int)
df['BirthDay'] = df['BirthDay'].astype(int)
```

```python
df.head(5)
```

```python
df.dtypes
```

##### Converting e-mail to get the domain and use label encoder

```python
df['domain'] = df['E-mail'].apply(lambda x: x.split('@')[1])
```

```python
le = LabelEncoder()
```

```python
df['domain_encoded'] = le.fit_transform(df['domain'])
```

```python
print(df['domain_encoded'])
```

```python
df.head(5)
```

```python
df.drop('E-mail', axis=1, inplace=True)
```

##### Converting countries with label encoder

```python
df['country_encoded'] = le.fit_transform(df['Country'])
```

```python
df.head(5)
```

##### Converting RegistrationDate to three columns

```python
df[['RegYear', 'RegMonth', 'RegDay']] = df['RegistrationDate'].str.split('-', expand=True)
df.drop('RegistrationDate', axis=1, inplace=True)
```

```python
df['RegYear'] = df['RegYear'].astype(int)
df['RegMonth'] = df['RegMonth'].astype(int)
df['RegDay'] = df['RegDay'].astype(int)
```

##### Converting LastPayment to three columns

```python
df[['LPYear', 'LPMonth', 'LPDay']] = df['LastPayment'].str.split('-', expand=True)
df.drop('LastPayment', axis=1, inplace=True)
```

```python
df['LPYear'] = df['LPYear'].astype(int)
df['LPMonth'] = df['LPMonth'].astype(int)
df['LPDay'] = df['LPDay'].astype(int)
```

##### Converting Payment Methods with Label Encoder

```python
df['pm_encoded'] = le.fit_transform(df['PaymentMethod'])
```

##### Converting Primary Device with Label Encoder

```python
df['pd_encoded'] = le.fit_transform(df['PrimaryDevice'])
```

##### Converting Other Device with Label Encoder

```python
df['od_encoded'] = le.fit_transform(df['OtherDevice'])
```

### Now we can start building some models

```python
X = df.drop(["SubscriptionStatus", "ID", "FullName","Country","PaymentMethod","PrimaryDevice","OtherDevice","domain"], axis=1)
y = df["SubscriptionStatus"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=1000)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)
```

### Let's check if we use 70/30 split instead how it will affect the result

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
clf = MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=1000)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)
```

#### Since the difference is only about 2%, I'd say that the dataset is not exactly sensitive on the training vs testing set size. After this I want to build some more models and see if we could aim to get something closer to this number

```python
clf = DecisionTreeClassifier(random_state=3)
clf.fit(X, y)
```

```python
importances = clf.feature_importances_
```

```python
for feature, importance in zip(X.columns, importances):
    print(f'{feature}: {importance}')
```
